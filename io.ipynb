```python
import networkx as nx
import matplotlib.pyplot as plt

# Definir as etapas do fluxo de trabalho e suas relaes,
# com quebra de pargrafo para termos com mais de 2 palavras e
# especificando o contexto de aprendizagem de graduandos em Engenharia de Software
stages = {
    "Coleta de Dados\n(Cdigos Python\nde Graduandos)": [],
    "Pr-processamento\nde Cdigo": ["Coleta de Dados\n(Cdigos Python\nde Graduandos)"],
    "Anlise de Cdigo com\nLLM (Identificao\nde Erros e Complexidade)": ["Pr-processamento\nde Cdigo"],
    "Extrao de Mtricas\nde Cdigo (Complexidade\nCiclomtica e Acoplamento)": ["Anlise de Cdigo com\nLLM (Identificao\nde Erros e Complexidade)"],
    "Identificao de Padres\nde Erros e Complexidade": ["Extrao de Mtricas\nde Cdigo (Complexidade\nCiclomtica e Acoplamento)"],
    "Desenvolvimento de\nEstratgias de Refatorao\n(com LLMs)": ["Identificao de Padres\nde Erros e Complexidade"],
    "Aplicao de\nRefatorao\nem Cdigo": ["Desenvolvimento de\nEstratgias de Refatorao\n(com LLMs)"],
    "Avaliao da\nRefatorao (Reduo de\nComplexidade e Acoplamento)": ["Aplicao de\nRefatorao\nem Cdigo"],
    "Feedback para\nGraduandos (com LLMs)": ["Avaliao da\nRefatorao (Reduo de\nComplexidade e Acoplamento)"],
    "Anlise da\nEfetividade da\nAprendizagem": ["Feedback para\nGraduandos (com LLMs)"],
    "Publicao\nAcadmica": ["Anlise da\nEfetividade da\nAprendizagem"],
}

# Funo para criar e desenhar o grafo direcionado (mantendo a mesma estrutura da funo original)
def draw_workflow_graph(stages):
    G = nx.DiGraph()

    # Adicionar ns e arestas ao grafo
    for stage, dependencies in stages.items():
        G.add_node(stage)
        for dependency in dependencies:
            G.add_edge(dependency, stage)

    # Definir o layout do grafo com espaamento
    pos = nx.spring_layout(G, seed=62, k=3.3)

    # Definir cores pastel escuras (prximas ao preto) para os ns,
    # agrupando por similaridade de etapas
    color_map = {
        "Coleta de Dados\n(Cdigos Python\nde Graduandos)": "#4B3C4D",
        "Pr-processamento\nde Cdigo": "#4B3C4D",
        "Anlise de Cdigo com\nLLM (Identificao\nde Erros e Complexidade)": "#3C4B5A",
        "Extrao de Mtricas\nde Cdigo (Complexidade\nCiclomtica e Acoplamento)": "#3C4B5A",
        "Identificao de Padres\nde Erros e Complexidade": "#5A3C4B",
        "Desenvolvimento de\nEstratgias de Refatorao\n(com LLMs)": "#5A3C4B",
        "Aplicao de\nRefatorao\nem Cdigo": "#3C5A4B",
        "Avaliao da\nRefatorao (Reduo de\nComplexidade e Acoplamento)": "#3C5A4B",
        "Feedback para\nGraduandos (com LLMs)": "#4B5A3C",
        "Anlise da\nEfetividade da\nAprendizagem": "#4B5A3C",
        "Publicao\nAcadmica": "#4B5A3C",
    }

    # Desenhar os ns com estilo e cores
    nx.draw_networkx_nodes(G, pos, node_size=2900, node_color=[color_map[node] for node in G.nodes()],
                           linewidths=2, edgecolors="black", node_shape='o')

    # Desenhar as arestas direcionadas com estilo
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, edge_color="darkgray",
                           arrowsize=12, arrowstyle='->', connectionstyle='arc3,rad=0.18')

    # Adicionar rtulos aos ns com estilo
    nx.draw_networkx_labels(G, pos, font_size=6, font_family="sans", font_weight="bold",
                            font_color="white")

    # Adicionar ttulo e remover eixos, especificando o contexto
    plt.title("Fluxo de Trabalho Acadmico:\nLLMs, Anlise de Cdigo e Refatorao", fontsize=12, fontweight="bold", color="white")
    plt.axis("off")

    # Definir fundo preto
    plt.gcf().set_facecolor("black")

    # Ajustar o tamanho da figura
    plt.figure(figsize=(36, 16))

    # Mostrar a visualizao
    plt.show()

# Executar a funo para desenhar o grafo
if __name__ == "__main__":
    draw_workflow_graph(stages)
```
