```python
import networkx as nx
import matplotlib.pyplot as plt

# Definir as etapas do fluxo de trabalho e suas relações,
# com quebra de parágrafo para termos com mais de 2 palavras e
# especificando o contexto de aprendizagem de graduandos em Engenharia de Software
stages = {
    "Coleta de Dados\n(Códigos Python\nde Graduandos)": [],
    "Pré-processamento\nde Código": ["Coleta de Dados\n(Códigos Python\nde Graduandos)"],
    "Análise de Código com\nLLM (Identificação\nde Erros e Complexidade)": ["Pré-processamento\nde Código"],
    "Extração de Métricas\nde Código (Complexidade\nCiclomática e Acoplamento)": ["Análise de Código com\nLLM (Identificação\nde Erros e Complexidade)"],
    "Identificação de Padrões\nde Erros e Complexidade": ["Extração de Métricas\nde Código (Complexidade\nCiclomática e Acoplamento)"],
    "Desenvolvimento de\nEstratégias de Refatoração\n(com LLMs)": ["Identificação de Padrões\nde Erros e Complexidade"],
    "Aplicação de\nRefatoração\nem Código": ["Desenvolvimento de\nEstratégias de Refatoração\n(com LLMs)"],
    "Avaliação da\nRefatoração (Redução de\nComplexidade e Acoplamento)": ["Aplicação de\nRefatoração\nem Código"],
    "Feedback para\nGraduandos (com LLMs)": ["Avaliação da\nRefatoração (Redução de\nComplexidade e Acoplamento)"],
    "Análise da\nEfetividade da\nAprendizagem": ["Feedback para\nGraduandos (com LLMs)"],
    "Publicação\nAcadêmica": ["Análise da\nEfetividade da\nAprendizagem"],
}

# Função para criar e desenhar o grafo direcionado (mantendo a mesma estrutura da função original)
def draw_workflow_graph(stages):
    G = nx.DiGraph()

    # Adicionar nós e arestas ao grafo
    for stage, dependencies in stages.items():
        G.add_node(stage)
        for dependency in dependencies:
            G.add_edge(dependency, stage)

    # Definir o layout do grafo com espaçamento
    pos = nx.spring_layout(G, seed=62, k=3.3)

    # Definir cores pastel escuras (próximas ao preto) para os nós,
    # agrupando por similaridade de etapas
    color_map = {
        "Coleta de Dados\n(Códigos Python\nde Graduandos)": "#4B3C4D",
        "Pré-processamento\nde Código": "#4B3C4D",
        "Análise de Código com\nLLM (Identificação\nde Erros e Complexidade)": "#3C4B5A",
        "Extração de Métricas\nde Código (Complexidade\nCiclomática e Acoplamento)": "#3C4B5A",
        "Identificação de Padrões\nde Erros e Complexidade": "#5A3C4B",
        "Desenvolvimento de\nEstratégias de Refatoração\n(com LLMs)": "#5A3C4B",
        "Aplicação de\nRefatoração\nem Código": "#3C5A4B",
        "Avaliação da\nRefatoração (Redução de\nComplexidade e Acoplamento)": "#3C5A4B",
        "Feedback para\nGraduandos (com LLMs)": "#4B5A3C",
        "Análise da\nEfetividade da\nAprendizagem": "#4B5A3C",
        "Publicação\nAcadêmica": "#4B5A3C",
    }

    # Desenhar os nós com estilo e cores
    nx.draw_networkx_nodes(G, pos, node_size=2900, node_color=[color_map[node] for node in G.nodes()],
                           linewidths=2, edgecolors="black", node_shape='o')

    # Desenhar as arestas direcionadas com estilo
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, edge_color="darkgray",
                           arrowsize=12, arrowstyle='->', connectionstyle='arc3,rad=0.18')

    # Adicionar rótulos aos nós com estilo
    nx.draw_networkx_labels(G, pos, font_size=6, font_family="sans", font_weight="bold",
                            font_color="white")

    # Adicionar título e remover eixos, especificando o contexto
    plt.title("Fluxo de Trabalho Acadêmico:\nLLMs, Análise de Código e Refatoração", fontsize=12, fontweight="bold", color="white")
    plt.axis("off")

    # Definir fundo preto
    plt.gcf().set_facecolor("black")

    # Ajustar o tamanho da figura
    plt.figure(figsize=(36, 16))

    # Mostrar a visualização
    plt.show()

# Executar a função para desenhar o grafo
if __name__ == "__main__":
    draw_workflow_graph(stages)
```
